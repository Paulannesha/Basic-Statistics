```python
from scipy import stats
x=data["Volume"]
mu=data['Volume'].mean()
sigma=data['Volume'].std()
data['z_score']=(x-mu)/sigma
print(data['z_score'])
```

```python
#Show before and after comparisons of the data distributions
```

```python
import pandas as pd 
import numpy as np
d1=data['Volume']
df=pd.DataFrame(d1)
means=df.mean()
stds=df.std()
standardized_df = (df - means) / stds
print("Original data:")
print(df)
print("\nStandardized data:")
print(standardized_df)
```
